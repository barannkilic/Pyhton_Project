# GOREV 2
### SaleCheckInDayDiff değişkenini EB_Score adında kategorik bir değişkene çeviriniz.

```python
bins = (0, 7, 30, 90, df["SaleCheckInDayDiff"].max())
labels = ("Last Minuters", "Potential Planners", "Planners", "Early Bookers")
```

```python
df["EB_Score"] = pd.cut(df["SaleCheckInDayDiff"], bins, labels=labels)
df.head()
```

# GOREV 4
### City-Concept-Seasons kırılımının cıktısını "price"a gore sıralayanız.
### Cıktıyı agg_df olarak kaydet

```python
df.head()
```

```python
agg_df = df.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price", ascending=False)
agg_df.head(10)
```

# GOREV 5
### İndekste yer alan isimleri degiskene ceviriniz.
### Ucuncu sorunun cıktısında yer alan PRICE dısındakı tum degiskenler index isimleridir. Bu isimleri Degisken isimlerine ceviriniz

```python
agg_df.reset_index(inplace=True)
agg_df.head()
```

# GOREV 6
### Yeni seviye tabanlı müsterileri(persona) tanımlayınız.
### Yeni seviye tabanlı satısları tanımlayınız ve veri setine degisken olarak ekleyiniz.
### Yeni eklenecek degiskenin adı: sales_level_based
### Oncekı soruda elde edecegınız cıktıdakı gozlemlerı bir araya getirerek sales_level_based degiskenini olusturmanız gerekmektedir.

```python
agg_df.head()
```

```python
agg_df['sales_level_based'] = agg_df[["SaleCityName", "ConceptName", 
                                      "Seasons"]].agg(lambda x: '_'.join(x).upper(), axis=1)
```

```python
agg_df.head(10)
```

# GOREV 7 
### Yeni musterileri segmentlere ayırınız

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["Price"], 4, labels=["D", "C", "B", "A"])
agg_df.head(30)
agg_df.groupby("SEGMENT").agg({"Price": ["mean", "max", "sum"]})
```
